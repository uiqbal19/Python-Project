I attempted to clean the above graph by instead using the mean value by each year

```python
#Calculate the mean sales per year using the groupby function
mean_sales_by_year = df.groupby("YEAR_ID")["SALES"].mean()

#Create a line plot
plt.plot(mean_sales_by_year.index, mean_sales_by_year)

#Set labels and title
plt.xlabel("Year")
plt.ylabel("Mean sales")
plt.title("Mean sales over time")

#show the plot
plt.show()
```

**Based on the 2 plots above, we can see that the advantage of seaborn as opposed to matplotlib is that seaborn is taking the averages of the sales from each year whereas matplotlib is plotting the raw data. # **4. Data Analysis**

**4.1: Analysing sales based on region**

```python
#creating a new variable that groups the unique territory values and finding the sum of sales. After that, it resets the table - e.g. moving territory names back into a regular column
sales_by_region = df.groupby("TERRITORY")["SALES"].sum().reset_index()

#changing the size
plt.figure(figsize = (12, 6))

#making a barplot to analyse sales by region
sns.barplot(x = "SALES", y = "TERRITORY", data = sales_by_region, palette = "viridis")

#setting the title, axis labels and showing the graph
plt.title("Sales by region")
plt.xlabel("Total sales")
plt.ylabel("Region (Territory)")
plt.show()
```

**4.1: Analysing sales by country**



```python
#creating a new variable that groups the unique territory values and finding the sum of sales. After that, it resets the table - e.g. moving territory names back into a regular column
sales_by_region = df.groupby("COUNTRY")["SALES"].sum().reset_index()

#changing the size
plt.figure(figsize = (12, 6))

#making a barplot to analyse sales by region
sns.barplot(x = "SALES", y = "COUNTRY", data = sales_by_region, palette = "viridis")

#setting the title, axis labels and showing the graph
plt.title("Sales by country")
plt.xlabel("Total sales")
plt.ylabel("Country)")
plt.show()
```

**4.1: Analysing deal size by region**

```python
pivot_table_status = df.pivot_table(
    index='DEALSIZE',
    columns= ['TERRITORY','YEAR_ID'],
    values='QUANTITYORDERED',
    aggfunc='sum'
)

pivot_table_status.plot(kind='bar', figsize=(10,6))

plt.xlabel('DealSize', fontsize=12)

plt.ylabel('Total Orders', fontsize=12)

plt.title("Total order per region and dealsize", fontsize=15)

plt.xticks(rotation=0)

plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')

plt.show()
```

**4.1: Analysing by date**

```python
#creating a new variable that groups the unique date values and finding the sum of sales. After that, it resets the table
sales_by_product = df.groupby("YEAR_ID")["SALES"].sum().reset_index()

#changing the size
plt.figure(figsize = (12, 6))

#making a barplot to analyse sales by region
sns.lineplot(x = "YEAR_ID", y = "SALES", data = sales_by_product)

#setting the title, axis labels and showing the graph
plt.title("Sales by time")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()
```

Interestingly, we notice that when looking at sales in this matter (i.e. using a groupby as the data source in the graph), sales seem to be declining after 2004. This may be a discrepancy in how the graph had been plotted using the groupby, and would be worth investigating further.

**4.2: Percentage of sales by each region**

```python
# Calculate total sales
total_sales = df["SALES"].sum()

# Calculate percentage contribution
df["Percentage"] = (df["SALES"] / total_sales) * 100

# Create a bar plot
plt.figure(figsize = (10, 6))
sns.barplot(x = "Percentage", y = "TERRITORY", data = df, palette = "viridis")

# Set labels and title
plt.xlabel("Percentage of total sales")
plt.ylabel("Region (Territory)")
plt.title("Sales contribution by region")

# Show the plot
plt.show()
```

While the percentages don't make much sense in terms of contribution (e.g. North America having the highest contribution but only being slightly above 0.035%), an alternative way of representing this would be to multiply by 1000, 10000 etc, but the actual formula of calculating the percentages is correct

**4.3: Performance of products in terms of deal size**

In this, I calculated the mean quantity ordered

```python
# Create a pivot table with mean quantity ordered
pt_performance = df.pivot_table(
    index="DEALSIZE",
    columns="PRODUCTLINE",
    values="QUANTITYORDERED",
    aggfunc="mean"
)

# Create a bar plot
plt.figure(figsize=(10, 6))
pt_performance.plot(kind="bar", figsize=(10, 6))

# Set labels and title
plt.xlabel("Deal Size", fontsize=12)
plt.ylabel("Total Orders", fontsize=12)
plt.title("Performance of products by deal size", fontsize=15)
plt.xticks(rotation=0)

# Add legend outside the plot
plt.legend(bbox_to_anchor=(1.02, 1), loc="upper left")

# Show the plot
plt.show()
```

**4.6: Calculating the percentage of sales by top-performing products**

```python
#Calculate total sales for each product line
total_sales = df.groupby("PRODUCTLINE")["SALES"].sum()
print(total_sales)

#Calculate total sales for all products
all_sales = df["SALES"].sum()
print(all_sales)

#Calculate percentage of sales for each product line
percentage = (total_sales / all_sales) * 100

#Add percentages to df using the .map function
df['percentage'] = df['PRODUCTLINE'].map(percentage)

#Plotting the bar plot
plt.figure(figsize=(12, 6))

#Create a bar plot with hue (color) based on product line
sns.barplot(x = "PRODUCTLINE", y = "percentage", data = df, palette = "viridis")

#Set labels and title
plt.xlabel("Product")
plt.ylabel("Percentage of total sales")
plt.title("Percentage of sales by product")

#Show the plot
plt.show()
```

# 1. Summary of findings and insights from the analysis

## To summarise the findings, we can see that from the graph in section 3.1, from 2003 to 2004, there was a solid increase in the amount of sales, with classic cars, trucks and buses accounting for the vast majority of sales. We can also see that all regions performed approximately equally during this time in terms of their sales. However, from 2004 to 2005 we see a sharp increase in sales. This could be due to increased demand of those products, an upturn in overall economic conditions or more attractive offers and sales being offered. 


## Something I've noticed in the graph in section 4.3 (performance of products by deal size) is that the total orders of planes performs the best out of any other product category when in a large deal size. Based on this, it would make sense to discount planes or to even explore other product lines as part of large deals in order to increase sales. However, a caveat to this is that the total orders of ships do not seem to benefit from large deal sizes, and so they can be excluded from any such promotions.
